```python
import sys
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('2017_1.csv')
data2 = pd.read_csv('2017_2.csv')
data3 = pd.read_csv('2018_1.csv')
data4 = pd.read_csv('2018_2.csv')
data5 = pd.read_csv('2019_1.csv')
data6 = pd.read_csv('2019_2.csv')

df1 = pd.DataFrame(data1, dtype=object)
df2 = pd.DataFrame(data2, dtype=object)
df3 = pd.DataFrame(data3, dtype=object)
df4 = pd.DataFrame(data4, dtype=object)
df5 = pd.DataFrame(data5, dtype=object)
df6 = pd.DataFrame(data6, dtype=object)
```

```python
total_df = pd.concat([df1, df2, df3, df4, df5, df6])
```

```python
def keyword_diff(keyword, df):
    diff = df[df['body'].str.contains(keyword, case=False, regex=True) & ~total_df['body'].str.contains("climate change|global warming|greenhouse effect|greenhouse gas*", case=False, regex=True)] 
    print("Number of articles containing", keyword, "and not other chosen keywords: ", len(diff))
    return diff


#diff = keyword_diff("Paris Agreement", total_df)
def keyword_filter(df):
    filtered_df = df[df['body'].str.contains("climate change|global warming|greenhouse effect|greenhouse gas*", regex=True, case=False)]
    print("Number of articles containing the keywords are: ", len(filtered_df))
    return filtered_df
```

```python

filtered_df = keyword_filter(total_df)
#dfs = filter_df("climate change|global warming|greenhouse effect|climate crisis|climate emergency", [df1, df2, df3, df4, df5, df6] )
#dfs

```
