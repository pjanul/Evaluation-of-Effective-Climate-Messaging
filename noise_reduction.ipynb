```python
import pandas as pd 
import numpy as np
import re
```

```python
names = ['id', 'date_publish', 'outlet', 'headline', 'lead', 'body', 'authors', 'domain', 'url', 'political_leaning']
data = pd.read_csv("climate_data.csv",encoding='latin1',  usecols=names)
df = pd.DataFrame(data)
```

```python
df['body'] = df['body'].astype(str)
df['id'] = df['id'].astype(int)
clean_df = df.copy()
```

```python
#clean_df.to_csv("test_climate_data.csv")
```

```python


breitbart_df = df[(df.outlet == "Breitbart")]
breitbart_df['body'] = breitbart_df['body'].str.replace("SIGN UP FOR OUR NEWSLETTER", "")


fox_df = df[(df.outlet == "Fox News")]
fox_df['body'] = fox_df['body'].str.replace("\*\*Want FOX News Halftime Report in your inbox every day? Sign up here.\*\*", "")

nyt_df = df[(df.outlet == "The New York Times")]
nyt_df['body'] = nyt_df['body'].str.replace(r'Advertisement Continue reading the main story', '', regex=True)
nyt_df['body'] = nyt_df['body'].str.replace(r'Photo\n', '', regex=True)
nyt_df['body'] = nyt_df['body'].str.replace(r'Newsletter Sign Up .* View all New York Times newsletters', '', regex=True)

nbc_df = df[(df.outlet == "NBC News")]
nbc_df['body'] = nbc_df['body'].str.replace(r'Play Facebook\nTwitter\nGoogle Plus', '', regex=True)

slate_df = df[(df.outlet == "Slate")]
slate_df['body'] = slate_df['body'].str.replace(r'Advertisement\n', '', regex=True)
slate_df['body'] = slate_df['body'].str.replace(r'Get Slate in your inbox.', '', regex=True)

npr_df = df[(df.outlet == "NPR")]
npr_df['body'] = npr_df['body'].str.replace(r'Enlarge this image toggle.*\n', '', regex=True)

usa_today_df = df[(df.outlet == "USA Today")]
usa_today_df['body'] = usa_today_df['body'].str.replace(r'Read or Share this story:', '', regex=True)
usa_today_df['body'] = usa_today_df['body'].str.replace(r'NEWSLETTERS Get the newsletter delivered to your inbox .* More newsletters', '', regex=True)
#npr_df = npr_df[nprdf.isin(['refused']).any(axis=1)]
```

```python
nyt_df.to_csv("test_climate_data.csv")
```

```python
clean_df = df.copy()
print(clean_df)
#print(clean_df.head())
fixed_outlet_names = ["Breitbart", "Fox News", "Slate", "NPR", "The New York Times", "NBC News"]
fixed_outlet_dfs = [breitbart_df, fox_df, slate_df, npr_df, nyt_df, nbc_df]

for outlet in fixed_outlet_names:
    clean_df = clean_df[(clean_df['outlet'] != outlet)]
    
for outlet in fixed_outlet_dfs: 
    #print(outlet.head())
    clean_df = pd.concat([clean_df, outlet])
    

    
clean_df = clean_df.sort_values(by=["date_publish"])
clean_df
```
